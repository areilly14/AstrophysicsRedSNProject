## Define wavelengths and range

```python
import numpy as np
from astropy.modeling.models import BlackBody
import astropy.units as u
import matplotlib.pyplot as plt
from dust_extinction.parameter_averages import G23
from dust_extinction.parameter_averages import F19
from dust_extinction.parameter_averages import F99
import sncosmo
import pandas as pd
from bayesn import SEDmodel

plt.rcParams.update({'font.size': 15})

# wavelengths and spectrum are 1D arrays
# wavelengths between 1000 and 30000 A
wavelengths = np.logspace(np.log10(0.1), np.log10(10), num=1000)*u.micron

# initialize the model
ext1 = G23(Rv=4)
ext2 = G23(Rv=3.5)
ext3 = G23(Rv=3)
ext4 = G23(Rv=2.5)
ext5 = G23(Rv=2)

# get the extinguished blackbody flux for different amounts of dust
wave_ext1 = ext1(wavelengths)
wave_ext2 = ext2(wavelengths)
wave_ext3 = ext3(wavelengths)
wave_ext4 = ext4(wavelengths)
wave_ext5 = ext5(wavelengths)

# plot the intrinsic and extinguished fluxes
fig, ax = plt.subplots()

ax.plot(wavelengths, wave_ext1, "r", label='$R(V) = 4$')
ax.plot(wavelengths, wave_ext2, "y", label='$R(V) = 3.5$')
ax.plot(wavelengths, wave_ext3, "g", label='$R(V) = 3$')
ax.plot(wavelengths, wave_ext4, "c", label='$R(V) = 2.5$')
ax.plot(wavelengths, wave_ext5, "b", label='$R(V) = 2$')

ax.set_xlabel('$\lambda$ [$\mu m$]')
ax.set_ylabel('$A_{\lambda}/A_{V}$')

ax.set_title('Extinction Curves with G23')
ax.set_ylim(0,5)
ax.set_xlim(0,1.5)

ax.legend(loc='best', fontsize = 15)
plt.tight_layout()
fig.savefig("/root/partiiiproject/ExtinctionCurvePlots/DustCurveG23.pdf", bbox_inches='tight')
plt.show()
```
