# Plottiing graphs using all the Supernova but this time a lot better

# Load modules

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sncosmo
from bayesn import SEDmodel
import os
import pandas as pd
import math
from scipy import stats
```

## Generate a list of all SNe 

```python
directory = os.fsencode("/root/partiiiproject/Archive")
supernovae_names = ["sn2006hb"]
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".dat"): 
        supernovae_names.append(filename[:-4])
        continue
    else:
        continue
```

## Declare dust laws

```python
dust_laws = ("F99", "F19", "G23")
```

## Load model

```python
model_name = ""
#model_name = "W22_F19_model"
```

```python


filter_yaml = "/root/partiiiproject/bayesn-filters/bayesn-filters/filters.yaml"

if model_name == "":
    model = SEDmodel(load_model="W22_model", filter_yaml = filter_yaml)
else:
    model = SEDmodel(load_model=model_name, filter_yaml = filter_yaml)
```

## Plotting Redshift and Distance

```python

column_names = ("F99_x", "F99_y", "F99_x_std", "F99_y_std", "F19_x", "F19_y", "F19_x_std", "F19_y_std", "G23_x", "G23_y", "G23_x_std", "G23_y_std",)

df = pd.DataFrame(columns=column_names)
df = pd.DataFrame(df, index=supernovae_names)

for supernova_name in supernovae_names:
    #Get original SN data file
    supernova_file = "/root/partiiiproject/Archive/"+supernova_name+".dat"
    meta, lcdata = sncosmo.read_snana_ascii(supernova_file, default_tablename='OBS')
    z = meta['REDSHIFT_FINAL']

    for dust_law in dust_laws:
        sampleSummaryDir = "DustLawsFitOfAllSamples"+model_name+"/"+supernova_name + dust_law + "_fit_summary.csv"
        try:
            samplesSummary = pd.read_csv(sampleSummaryDir)
            samplesSummary = samplesSummary.rename(index=samplesSummary['Unnamed: 0'])
            df.at[supernova_name, dust_law+"_x"] = z
            df.at[supernova_name, dust_law+"_y"] = samplesSummary.loc['mu[0]', 'mean']
            df.at[supernova_name, dust_law+"_y_std"] = samplesSummary.loc['mu[0]', 'sd']
        except FileNotFoundError:
            continue

#Plot the actual graph
cs = ["g", "r", "b"]
for i, dust_law in enumerate(dust_laws):
    plt.errorbar(df[dust_law+"_x"], df[dust_law+"_y"], df[dust_law+"_y_std"], fmt=cs[i]+'x', label=dust_law)

#Properties of the plot and display
plt.xlabel('redshift z')
plt.ylabel('mu')
plt.legend()
plt.savefig("/root/partiiiproject/DustLawsGeneralGraphs/distanceredshiftG23vsF99"+model_name+""+".png", bbox_inches='tight') 
plt.show()
```

## Plotting a graph of Rv and Av Values

```python

column_names = ("F99_x", "F99_y", "F99_x_std", "F99_y_std", "F19_x", "F19_y", "F19_x_std", "F19_y_std", "G23_x", "G23_y", "G23_x_std", "G23_y_std",)

df = pd.DataFrame(columns=column_names)
df = pd.DataFrame(df, index=supernovae_names)
dust_laws = ("F99", "F19", "G23")

for supernova_name in supernovae_names:
    
    #Get original SN data file
    supernova_file = "/root/partiiiproject/Archive/"+supernova_name+".dat"
    meta, lcdata = sncosmo.read_snana_ascii(supernova_file, default_tablename='OBS')
    z = meta['REDSHIFT_FINAL']

    if supernova_name == "sn2002cv":
        continue

    for dust_law in dust_laws:
        sampleSummaryDir = "DustLawsFitOfAllSamples"+model_name+"/"+supernova_name + dust_law + "_fit_summary.csv"
        try:
            samplesSummary = pd.read_csv(sampleSummaryDir)
            samplesSummary = samplesSummary.rename(index=samplesSummary['Unnamed: 0'])
            df.at[supernova_name, dust_law+"_x"] = samplesSummary.loc['AV[0]', 'mean']
            df.at[supernova_name, dust_law+"_x_std"] = samplesSummary.loc['AV[0]', 'sd']
            df.at[supernova_name, dust_law+"_y"] = samplesSummary.loc['RV[0]', 'mean']
            df.at[supernova_name, dust_law+"_y_std"] = samplesSummary.loc['RV[0]', 'sd']
        except FileNotFoundError:
            continue

#Plot the actual graph
cs = ["g", "r", "b"]

df = df.iloc[2:]
print(df)
#Clean up
df = df.dropna(axis=1, how = "all")
df = df.dropna(axis =0, how = "any")
print(df)

for i, dust_law in enumerate(dust_laws):
    plt.plot(df[dust_law+"_x"], df[dust_law+"_y"], cs[i]+'x', label=dust_law)
    b, a = np.polyfit(df[dust_law+"_x"].astype(float), df[dust_law+"_y"].astype(float), deg=1)
    plt.plot(df[dust_law+"_x"].astype(float), a + b * df[dust_law+"_x"].astype(float), color=cs[i], lw=1, alpha =0.7)

#Properties of the plot and display
plt.xlabel(f'$A_V$',fontsize = 20)
plt.ylabel(f'$R_V$',fontsize = 20)
#plt.xlim([0,3])
#plt.ylim([0,6])
plt.title(f"$R_V$ and $A_V$ for F99, F19 and G23", fontsize = 20)
plt.legend(fontsize = 20)
plt.savefig("/root/partiiiproject/DustLawsGeneralGraphs/RvAvAllSNe"+model_name+".png", bbox_inches='tight') 
plt.show()
```

```python

column_names = ("F99_x", "F99_y", "F99_x_std", "F99_y_std", "F19_x", "F19_y", "F19_x_std", "F19_y_std", "G23_x", "G23_y", "G23_x_std", "G23_y_std",)

df = pd.DataFrame(columns=column_names)
df = pd.DataFrame(df, index=supernovae_names)
model_name = ""
for supernova_name in supernovae_names:

    if supernova_name== "sn2002cv":
        continue
    
    #Get original SN data file
    supernova_file = "/root/partiiiproject/Archive/"+supernova_name+".dat"
    meta, lcdata = sncosmo.read_snana_ascii(supernova_file, default_tablename='OBS')
    z = meta['REDSHIFT_FINAL']

    dust_laws = ["F99", "F19", "G23"]
    for i, dust_law in enumerate(dust_laws):
        sampleSummaryDir = "DustLawsFitOfAllSamples"+model_name+"/"+supernova_name + dust_law + "_fit_summary.csv"
            
        try:
            samplesSummary = pd.read_csv(sampleSummaryDir)
            samplesSummary = samplesSummary.rename(index=samplesSummary['Unnamed: 0'])
            df.at[supernova_name, dust_law+"_x"] = samplesSummary.loc['AV[0]', 'mean']
            df.at[supernova_name, dust_law+"_x_std"] = samplesSummary.loc['AV[0]', 'sd']
            df.at[supernova_name, dust_law+"_y"] = samplesSummary.loc['mu[0]', 'mean']
            df.at[supernova_name, dust_law+"_y_std"] = samplesSummary.loc['mu[0]', 'sd']
        except FileNotFoundError:
            continue


#Plot the actual graph
cs = ["g", "r", "b"]

df = df.iloc[2:]

plt.plot([0,2.5], [0,0], color = 'k', alpha = 0.8, label = "F99 baseline")

#Clean up
df = df.dropna(axis=1, how = "all")
df =df.dropna(axis =0, how = "any")
print(df)

labels_display = ["F99","F19","G23"]

for i, dust_law in enumerate(dust_laws):

    if dust_law == "F99":
        continue
            
    plt.errorbar(df[dust_law+"_x"], df[dust_law+"_y"]-df["F99_y"], df[dust_law+"_y_std"].astype(float), fmt =cs[i]+'x', alpha =0.2)
    plt.plot(df[dust_law+"_x"], df[dust_law+"_y"]-df["F99_y"], cs[i]+'x', alpha=0.6)
    #b, a = np.polyfit(df[dust_law+"_x"].astype(float), df[dust_law+"_y"].astype(float)-df["F99_y"].astype(float), deg=1)
    #plt.plot(df[dust_law+"_x"].astype(float), a + b * df[dust_law+"_x"].astype(float), color=cs[i], lw=1)
    bin_means, bin_edges, binnumber = stats.binned_statistic(df[dust_law+"_x"].astype(float), df[dust_law+"_y"].astype(float)-df["F99_y"].astype(float), 'mean', bins=10)
    #plt.hlines(bin_means, bin_edges[:-1], bin_edges[1:], colors=cs[i], lw=4, label='binned mean '+ identity, alpha=0.9)
    print(binnumber)
    bin_mean_std = np.zeros(len(bin_means))
    for j in np.arange(len(bin_means)):
        c1 = bin_edges[j]<=df[dust_law+"_x"]
        c2 = df[dust_law+"_x"]<=bin_edges[j+1]
        df_edit = df.where(c1 & c2)
        #df_edit = df_edit.dropna(axis=1, how = "all")
        df_edit =df_edit.dropna(axis =0, how = "any")
        bin_mean_std[j] = np.sqrt(sum((df_edit[dust_law+"_y_std"].astype(float)**2)/len(df_edit[dust_law+"_y_std"])**2))
    
    print(bin_mean_std)
    plt.errorbar(0.5*(bin_edges[:-1]+bin_edges[1:]), bin_means, bin_mean_std, fmt=cs[i]+'x', alpha =1, capsize = 10, label=" "+ labels_display[i])


#Properties of the plot and display
plt.xlabel(f'$A_V$',fontsize = 20)
plt.ylabel(f'$\Delta$ $\mu$',fontsize = 20)
plt.xlim([0,2.7])
plt.ylim([-0.2,0.9])
plt.legend(fontsize = 15, loc = "upper left")
plt.title(f'Modulus distance $\mu$',fontsize = 20)
plt.savefig("/root/partiiiproject/DustLawsGeneralGraphs/distancedifferencesWithAVG23vsF99color"+model_name+".png", bbox_inches='tight') 
plt.show()
```

## Comparison of different models

```python

column_names = ("W22_x", "W22_y", "W22_x_std", "W22_y_std","F99_x", "F99_y", "F99_x_std", "F99_y_std", "F19_x", "F19_y", "F19_x_std", "F19_y_std", "G23_x", "G23_y", "G23_x_std", "G23_y_std",)

df = pd.DataFrame(columns=column_names)
df = pd.DataFrame(df, index=supernovae_names)

for supernova_name in supernovae_names:

    if supernova_name == "sn2002cv":
        continue
    
    #Get original SN data file
    supernova_file = "/root/partiiiproject/Archive/"+supernova_name+".dat"
    meta, lcdata = sncosmo.read_snana_ascii(supernova_file, default_tablename='OBS')
    z = meta['REDSHIFT_FINAL']

    for model_name in ["", "W22_F19_model", "W22_G23_model", "W22_F99_model"]:
        if model_name == "W22_F19_model":
            dust_law = "F19"
            identity = dust_law
        elif model_name == "":
            dust_law = "F99"
            identity = "W22"
        elif model_name == "W22_F99_model":
            dust_law = "F99"
            identity = dust_law
        elif model_name == "W22_G23_model":
            dust_law = "G23"
            identity = dust_law
        
        sampleSummaryDir = "DustLawsFitOfAllSamples"+model_name+"/"+supernova_name + dust_law + "_fit_summary.csv"
        try:
            samplesSummary = pd.read_csv(sampleSummaryDir)
            samplesSummary = samplesSummary.rename(index=samplesSummary['Unnamed: 0'])
            df.at[supernova_name, identity+"_x"] = samplesSummary.loc['AV[0]', 'mean']
            df.at[supernova_name, identity+"_x_std"] = samplesSummary.loc['AV[0]', 'sd']
            df.at[supernova_name, identity+"_y"] = samplesSummary.loc['mu[0]', 'mean']
            df.at[supernova_name, identity+"_y_std"] = samplesSummary.loc['mu[0]', 'sd']
        except FileNotFoundError:
            continue

#Plot the actual graph
cs = ["r", "b", "g", "k"]

#Clean up
df = df.dropna(axis=1, how = "all")
df =df.dropna(axis =0, how = "any")


for i, identity in enumerate(["F19", "G23", "F99"]):
    #plt.errorbar(df[dust_law+"_x"], df[dust_law+"_y"]-df["F99_y"], df[dust_law+"_y_std"].astype(float), fmt =cs[i]+'x', label=dust_law, alpha =0.3)
    plt.plot(df[identity+"_x"], df[identity+"_y"]-df["W22_y"], cs[i]+'x', label=identity)
    b, a = np.polyfit(df[identity+"_x"].astype(float), df[identity+"_y"].astype(float)-df["W22_y"].astype(float), deg=1)
    plt.plot(df[identity+"_x"].astype(float), a + b * df[identity+"_x"].astype(float), color=cs[i], lw=1)

    #Binned averages
    bin_means, bin_edges, binnumber = stats.binned_statistic(df[identity+"_x"].astype(float), df[identity+"_y"].astype(float)-df["W22_y"].astype(float), 'mean', bins=10)
    plt.hlines(bin_means, bin_edges[:-1], bin_edges[1:], colors=cs[i], lw=4, label='binned mean '+ identity, alpha=0.5)


#Properties of the plot and display
plt.xlabel(f'$A_V$',fontsize = 20)
plt.ylabel(f'$\Delta$ $\mu$',fontsize = 20)
plt.xlim([0,2.5])
#plt.ylim([-0.2,0.7])
plt.legend(fontsize = 10)
plt.title(f'$A_V$ and modulus distance $\mu$',fontsize = 20)
plt.savefig("/root/partiiiproject/DustLawsGeneralGraphs/distancedifferencesWithAVG23vsF99colorDifferentModels.png", bbox_inches='tight') 
plt.show()
    
```
